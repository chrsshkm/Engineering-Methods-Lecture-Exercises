## Trapezoidal Rule

### Algorithm:
1. Divide the interval [a, b] into n subintervals of equal width h = (b - a)/n.
2. Compute the sum: Integral ≈ (h/2) * [f(x₀) + 2f(x₁) + 2f(x₂) + ... + 2f(xₙ₋₁) + f(xₙ)].


## Simpson's Rule

### Algorithm:
1. Divide the interval [a, b] into an even number n of subintervals.
2. Compute the sum: Integral ≈ (h/3) * [f(x₀) + 4f(x₁) + 2f(x₂) + 4f(x₃) + ... + 4f(xₙ₋₁) + f(xₙ)].
3. Return the result.


```python
# Tour code here
import numpy as np

def simpsons_rule(f, a, b, n):
  
    if n % 2 != 0:
        raise ValueError("NOOOOO WRONG")

    h = (b - a) / n
    x = np.linspace(a, b, n + 1)     
    y = f(x)

   
    S = y[0] + y[-1]
    S += 4 * np.sum(y[1:-1:2])       
    S += 2 * np.sum(y[2:-1:2])        

    return (h / 3) * S


if __name__ == "__main__":
    f = lambda x: x**2
    result = simpsons_rule(f, 0, 1, n=100)  
    print("Integral ≈", result)  
```
